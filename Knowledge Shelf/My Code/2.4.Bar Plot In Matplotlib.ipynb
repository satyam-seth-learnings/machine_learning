```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

tick_label=['One','Two','Three','Four','Five']
plt.bar(x,y,tick_label=tick_label,width=0.7)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

tick_label=['One','Two','Three','Four','Five']
plt.bar(x,y,tick_label=tick_label,width=0.7,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

tick_label=['One','Two','Three','Four','Five']
plt.bar(x,y,tick_label=tick_label,width=0.7,color=['green','blue'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.show()
```
