```python
import matplotlib.pyplot as plt
import csv
```

```python
x=[]
y=[]

with open('2.9. Plot Data from CSV & Text Files using Matplotlib.csv','r') as csvfile:
    plots=csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
        
plt.plot(x,y,label='file')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Test Graph')

plt.legend()
plt.show()
```
