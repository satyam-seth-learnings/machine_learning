```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.arange(0.0,2.0,0.01)
s=1+np.cos(2*np.pi*t)
plt.plot(t,s)
plt.xlabel('Time(t)')
plt.ylabel('Voltage(ev)')
plt.title('Cosine Wave plot(Cos(x))')
plt.show()
```
