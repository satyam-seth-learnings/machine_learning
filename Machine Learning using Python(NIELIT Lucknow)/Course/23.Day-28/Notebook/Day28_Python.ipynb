```python
import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X=df.iloc[1:100,[3,4]].values
X.shape
```

```python
#Number of Clusters
k=5
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5)
km.fit(X)
y_predict=km.predict(X)
print(y_predict)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

```python
plt.scatter(X[:,0],X[:,1],c=y_predict,cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km.fit(X)
y_predict=km.predict(X)
print(y_predict)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_predict,cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(X)
y_predict=km.predict(X)
print(y_predict)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_predict,cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2)
km.fit(X)
y_predict=km.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_predict,cmap="rainbow")
```

```python
km.inertia_
```

```python
#How many numbers are clusters dataset should be divided?
#What is most suitable number for cluster formation?
# it can decided by looking at the cluster and problem in hand
# ELBOW METHOD is used to find optimum number of clusters

from sklearn.cluster import KMeans
E=[]
for i in range(2,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    E.append(km.inertia_)
print(E)
    
```

```python
plt.plot(range(2,11),E)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Intertia")
plt.show()
```
